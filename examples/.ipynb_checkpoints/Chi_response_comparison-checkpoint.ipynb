```python
import numpy as np
from pandas import read_csv

# from average_atom.core.average_atom_new import AverageAtom as AA
from average_atom.core.aa_types import AverageAtomFactory
from average_atom.core.misc import jacobi_relaxation, sor
from average_atom.core.grids import NonUniformGrid
from average_atom.core.physics import ThomasFermi, FermiDirac

import matplotlib.pyplot as plt

from hnc.hnc.constants import *
from hnc.hnc.misc import rs_from_n, n_from_rs

plt.rcParams.update({'font.size': 10})
plt.rcParams["figure.figsize"] = (5,3.5)
plt.rcParams["figure.dpi"] = 200
plt.rcParams['xtick.labelsize']=14
plt.rcParams.update({'font.size': 12, 'font.family': 'serif'})
plt.rcParams['legend.fontsize'] = 8

color_cycle = plt.rcParams['axes.prop_cycle'].by_key()['color']
```

```python
# Aluminum
name='Al'
Z, A = 13, 27 

ρ_solid = 2.699 #g/cc, the solid density of aluminum 
ρ = ρ_solid#8.1 # g/cc
ni_cc = ρ/(A*amu_to_AU*AU_to_g)
ni_AU = ni_cc*invcc_to_AU
rs = rs_from_n(ni_AU)

R  = 10*rs

Ti_eV = 1 
Te_eV = 1

#

Te_AU = Te_eV*eV_to_AU
Ti_AU = Ti_eV*eV_to_AU


# aa_ZJ  = AverageAtomFactory.create_model("ZJ_ISModel", Z, A, Ti_AU, Te_AU, rs, R, name=name, ignore_vxc=True, xc_type='KSDT', Npoints=2000, rmin=1e-4, N_stencil_oneside = 2)
aa_W  = AverageAtomFactory.create_model("ZJ_ISModel_W", Z, A, Ti_AU, Te_AU, rs, R,  Weizsacker_λ=1/5, name=name, ignore_vxc=False, xc_type='KSDT', Npoints=2000, rmin=1e-4, N_stencil_oneside = 2)
aa_W.nf_init = 1/aa_W.grid.xs**1.5*np.exp(-2*aa_W.grid.xs)
aa_W.nf_init *= Z/aa_W.grid.integrate_f(aa_W.nf_init )
aa_W.ne = aa_W.nf_init + 0*aa_W.nb_init
# aa_ZJ_cs  = AverageAtomFactory.create_model("ZJ_CSModel", Z, A, Ti_AU, Te_AU, rs, R, name=name, ignore_vxc=False, Npoints=2000, rmin=1e-4, N_stencil_oneside = 2)
# aa_CS  = AverageAtomFactory.create_model("TFStarret2014", Z, A, Ti_AU, Te_AU, rs, R, name=name, ignore_vxc=False, Npoints=2000, rmin=1e-4, N_stencil_oneside = 2)
```

```python
from average_atom.core.physics import χ0_Lindhard, G_SLFC, χ_Lindhard, χ_TF

fig, ax = plt.subplots()

aa_W.Te, aa_W.μ = 10.3*eV_to_AU, aa_W.EF
ax.plot(aa_W.iet.k_array/rs/aa_W.kF, -χ_Lindhard(aa_W.μ, aa_W.Te, aa_W.iet.k_array/rs, aa_W.kF, finite_T=False, LFC=False),color=color_cycle[0], linestyle='-', label=r"Lindhard ($T=0$, no LFC)")
ax.plot(aa_W.iet.k_array/rs/aa_W.kF, -χ_Lindhard(aa_W.μ, aa_W.Te, aa_W.iet.k_array/rs, aa_W.kF, finite_T=False, LFC=True) ,color=color_cycle[0], linestyle='--', label=r"Lindhard ($T=0$, LFC)")
ax.plot(aa_W.iet.k_array/rs/aa_W.kF, -χ_Lindhard(aa_W.μ, aa_W.Te, aa_W.iet.k_array/rs, aa_W.kF, finite_T=True, LFC=False) ,color=color_cycle[1], linestyle='-', label=r"Lindhard ($\theta={0:0.2f}$, no LFC)".format(aa_W.Te/aa_W.EF))
ax.plot(aa_W.iet.k_array/rs/aa_W.kF, -χ_Lindhard(aa_W.μ, aa_W.Te, aa_W.iet.k_array/rs, aa_W.kF, finite_T=True, LFC=True)  ,color=color_cycle[1], linestyle='--', label=r"Lindhard ($\theta={0:0.2f}$, LFC)".format(aa_W.Te/aa_W.EF))
ax.plot(aa_W.iet.k_array/rs/aa_W.kF, -χ_TF(aa_W.iet.k_array/rs, aa_W.kF), color=color_cycle[2], label="TF")

ax.set_ylabel(r"$-\chi^{\rm Lindhard}_{ee}$")
ax.set_xlabel(r"$k/k_F$")
ax.axvline(2,color='k',linestyle='--')
# ax.set_yscale('symlog', linthresh=1e-1)
ax.set_yscale('log')
ax.set_xscale('log')
ax.set_ylim(1e-3,0.5)
ax.set_xlim(0.1,20)
ax.grid(which='both')
ax.legend()
plt.tight_layout()
# plt.savefig("../media/Lindhard_finiteT_LFC_comparison_4.png")
```
